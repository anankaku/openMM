```python
import pandas as pd

# Load the smiles data
smiles_df = pd.read_csv('peptoidff_benchmark/data/smiles.csv')
display(smiles_df.head())

# Check if we can find the entry for 17AB2-8-A
peptoid_id = '17AB2-8-A'

# The dataframe structure has IDs as columns, not as rows in a 'Molecule ID' column
if peptoid_id in smiles_df.columns:
    # Access the column with the name '17AB2-8-A' and take the first value
    smiles_string = smiles_df[peptoid_id].iloc[0]
    print(f"Found SMILES for {peptoid_id}: {smiles_string}")
else:
    print(f"ID {peptoid_id} not found in columns. Available columns:")
    print(smiles_df.columns)
```

```python
import pandas as pd

# Load the capped smiles data
capped_df = pd.read_csv('peptoidff_benchmark/data/capped_smiles.csv')
display(capped_df.head())

# Check for 06AA1-9-A
target_id = '06AA1-9-A'
if target_id in capped_df.columns:
    full_smiles = capped_df[target_id].iloc[0]
    print(f"Full Capped SMILES for {target_id}: {full_smiles}")
else:
    print(f"{target_id} not found in capped_smiles.csv")
```
